# Data Wrangling

```python
df = df.iloc[2:]
```

```python
df.head()
```

```python
df['Adult'] = df['Adult'].astype('int64')
df['Motors'] = df['Motors'].astype('int64')
df['Animals'] = df['Animals'].astype('int64')
df['Services'] = df['Services'].astype('int64')
df['For Sale'] = df['For Sale'].astype('int64')
df['No Category'] = df['No Category'].astype('int64')
df['Jobs'] = df['Jobs'].astype('int64')
df['Adult Businesses'] = df['Adult Businesses'].astype('int64')
df['Events'] = df['Events'].astype('int64')
df['Property'] = df['Property'].astype('int64')
df['Totals'] = df['Totals'].astype('int64')
```

```python
df.info()
```

```python
df.rename(columns={'adcategory':'URL'}, inplace=True)
```

```python
df.head()
```

```python
df['No Category'].value_counts()
```

```python
df.shape
```

```python
df = df[df['No Category'] == 0]
```

```python
df.shape
```

```python
df.drop(columns='No Category', inplace=True)
```

```python
df.tail()
```

```python
def split_url(url):
    pattern = r'[&?]'
    split_list = [item for sublist in [re.split(pattern, item) for item in url.split('/')] for item in sublist]
    return [item for item in split_list if item]
```

```python
df['URL'] = df['URL'].apply(split_url)
```

```python
df.head()
```

```python
df['URL'] = df['URL'].apply(lambda x: x[4:] if len(x) > 3 else [])
```

```python
df.head(10)
```

```python
df['items'] = df['URL'].apply(len)
```

```python
df.head(200)
```
