```python
def split_url(url):
    pattern = r'[&?]'
    split_list = [item for sublist in [re.split(pattern, item) for item in url.split('/')] for item in sublist]
    return [item for item in split_list if item]
```

```python
df['URL_first_half'] = df['URL_first_half'].apply(split_url)
```

```python
def filter_items(items):
    filtered_items = [item for item in items if not item.isupper()]
    filter_code = [item for item in items if item.isupper()]
    return filtered_items, filter_code
```
